```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
try:
    iris=pd.read_csv("iris.csv")
except FileNotFoundError:
    print("Error: file is not found ensure its in same directory.")
```

```python
# display first 5 rows
iris.head()
```

```python
# check tata types and missing values
iris.info()
```

```python
# check missing values in each column
iris.isnull().sum()
```

```python
# handle missiong values
iris.dropna(inplace=True)
```

```python
# data statistics
iris.describe()
```

```python
#calculate mean of vlues grouped by variety
iris.groupby('variety').mean()
```

```python
#Data visualization
# Line chart
plt.figure(figsize=(10,4))
plt.plot(iris['sepal.length'],label='Sepal Length')
plt.title("Trend of Sepal Length Over Samples")
plt.xlabel("Sample Index")
plt.ylabel("Sepal length in (cm)")
plt.legend()
plt.grid(True)
plt.show()
# we can see how the sepal length icreases or flactuates 
```

```python
# Bar chat  Averge petal length per variety
plt.figure(figsize=(8,4))
iris.groupby('variety')['petal.length'].mean().plot(kind='bar', color='purple')
plt.title('Average Petal Length per Variety')
plt.xlabel('Variety')
plt.ylabel('Petal Length (cm)')
# plt.grid(True)
plt.show()
# the longest petals are virginica, followed by Versicolor and them Setosa
```

```python
# Histogram - sepal width separation
plt.figure(figsize=(8,4))
plt.hist(iris['sepal.width'],bins=10,color='lightblue', edgecolor='black')
plt.title("Distribution of sepal width.")
plt.xlabel("sepal width (cm)")
plt.ylabel("Frequency")
plt.show()
```

```python
# Scatter Plot - Sepal Length vs Petal Length
plt.figure(figsize=(8,4))
sns.scatterplot(data=iris, x='sepal.length', y='petal.length', hue='variety')
plt.title('Sepal Length vs Petal Length by Variety')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.legend()
plt.show()

#shows cluster of varieties how the petal length increase as the length of sepal increases
# 
```
